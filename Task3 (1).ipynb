# Analyzing the data using graphs and charts

## Visualizing the Categories in terms of their counts

```python
plt.figure(figsize = (10,5))
sns.countplot(x = 'Category', data = data, palette= ["#fc9272","#aabbc1", "#dcc0d2"])
plt.title('CATEGORIES')
plt.xticks(rotation = 90);
```

## Count Plots for various Categories vs Region

```python
plt.figure(figsize=(10,5))
plt.title('CATEGORIES VS REGION COUNT PLOT')
sns.countplot(x=data['Category'],hue=data['Region'],palette='flare')
plt.xticks()
```

```python
plt.figure(figsize=(10,5))
plt.title('SHIP MODE VS REGION COUNT PLOT')
sns.countplot(x=data['Ship Mode'],hue=data['Region'],palette='crest')
plt.xticks()
```

```python
plt.figure(figsize = (10,5))
sns.countplot(x = 'Segment', data = data, palette='cool_r')
plt.title('SEGMENT')
plt.xticks(rotation = 90);
```

```python
data[['Quantity']].plot.hist(bins = 20, title = 'Purchases Quantity Distribution', color="lightpink")
```

```python
plt.figure(figsize = (10,5))
sns.countplot(x = 'Sub-Category', data = data, palette= ["#aabbc1","#fc9272", "#caa0d2",'#abc012','#eabcde'])
plt.title('SUB-CATEGORIES')
plt.xticks(rotation = 90);
```
